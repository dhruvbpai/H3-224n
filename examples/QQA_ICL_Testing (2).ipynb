# GPT Permutation Visualizations

```python
%cd /content/drive/MyDrive/h3/'QQP Dataset'/'GPT Permute results'/
!ls
!pwd
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
sns.set()
def graph_hists(path, col, title):
  fnames=os.listdir(path)
  num_hists = len(fnames)
  colors=["r","b","g","y"]
  plt.tight_layout()
  plt.rcParams.update({'font.size': 20})
  fig = plt.figure(figsize=(18,11))

  fnames=sorted(fnames, key=lambda x: int(x.split("_")[1]))

  for i, fname in enumerate(fnames):
    fields = fname.split("_")
    size = fields[1]
    data = pd.read_csv(fname).dropna()
    plt.subplot(2,2,i+1)
    plt.hist(data[col], color=colors[i], alpha=0.5, bins = [0.1*i for i in range(15)])
    plt.xlabel("Crossentropy")
    plt.ylabel("Frequency")
    plt.title("%s-shot" % size)
  fig.suptitle(title)

graph_hists("./", "# Mean CrossEntropy", "GPT Crossentropy Distributions under Permutation (Means)")
```

```python
graph_hists("./", "Stdev CrossEntropy", "GPT Crossentropy Distributions under Permutation (Standard Deviations)")
```

# GPT Noise Visualizations

```python
%cd /content/drive/MyDrive/h3/'QQP Dataset'/'GPT Noise Results'/
!ls
```

```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import make_interp_spline
plt.style.use('seaborn-whitegrid')


def get_accs(path):
  fnames=[i for i in os.listdir(path) if i[-1]=="v"]
  fnames=sorted(fnames, key=lambda x: int(x.split("_")[1]))
  accs=[]
  sizes=[]
  for fname in fnames:
    data =  pd.read_csv(path+fname)
    acc = np.where(data[data.columns[0]]==data[data.columns[1]])[0].shape[0]/data.shape[0]
    size = int(fname.split("_")[1])
    accs.append(acc)
    sizes.append(size)
  
  return np.array(sizes), np.array(accs)

noisy_sizes, noisy_accs = get_accs("./")
normal_sizes, normal_accs = get_accs("../GPT Results/")
```

```python
type(noisy_accs)
```

```python
plt.plot(noisy_sizes, noisy_accs, label="noisy")
plt.plot(normal_sizes[2:], normal_accs[2:], label="no noise")
plt.title("Accuracy vs ICL Shot Size")
plt.ylabel("Accuracy")
plt.xlabel("Shot Size")
plt.legend()
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
def fit_spline(x,y, c1, c2, label):
  X_Y_Spline = make_interp_spline(x, y, k=2)
  X_ = np.linspace(x.min(), x.max(), 500)
  Y_ = X_Y_Spline(X_)
  plt.plot(x, y, c1, label=label)
  plt.plot(X_, Y_, '-', color=c2)

```

```python
fit_spline(noisy_sizes, noisy_accs, "or", "gray", "Noisy")
fit_spline(normal_sizes[2:], normal_accs[2:],"ob","gray", "Normal")
plt.legend()
plt.xlabel("Shot Size")
plt.ylabel("Accuracy")
plt.title("GPT Accuracy on ICL in Noisy vs Non-noisy Label Domains")
```

```python
from sklearn.metrics import f1_score

def get_f1(path):
  fnames=[i for i in os.listdir(path) if i[-1]=="v"]
  fnames=sorted(fnames, key=lambda x: int(x.split("_")[1]))
  f1s=[]
  sizes=[]
  for fname in fnames:
    data =  pd.read_csv(path+fname)
    size = int(fname.split("_")[1])
    # print(data[data.columns[0]], data[data.columns[1]])
    f1s.append(f1_score(data[data.columns[0]], data[data.columns[1]]))
    sizes.append(size)
  
  return np.array(sizes), np.array(f1s)

noisy_sizes, noisy_f1 = get_f1("./")
normal_sizes, normal_f1 = get_f1("../GPT Results/")
```

```python
get_f1("./")
```

```python
fit_spline(noisy_sizes, noisy_f1, "or", "gray", "Noisy")
fit_spline(normal_sizes[2:], normal_f1[2:],"ob","gray", "Normal")
plt.legend()
plt.xlabel("Shot Size")
plt.ylabel("F1 Score")
plt.title("GPT F1 Score on ICL in Noisy vs Non-noisy Label Domains")
```

```python

def ce_scaling(path):
  fnames=[ i for i in os.listdir(path) if i[-1]=="v"]
  fnames=sorted(fnames, key=lambda x: int(x.split("_")[1]))
  mean_ces=[]; std_ces=[]; sizes=[]
  for fname in fnames:
    data =  pd.read_csv(path+fname).dropna()
    size = int(fname.split("_")[1])
    ce_m = np.mean(data[" Cross-entropy"])
    ce_s = np.std(data[" Cross-entropy"])
    mean_ces.append(ce_m)
    std_ces.append(ce_s)
    
    sizes.append(size)
  return sizes, mean_ces, std_ces
sizes, mean_ces, std_ces = ce_scaling("./")


print(sizes, mean_ces, std_ces)
plt.plot(sizes, mean_ces)
plt.errorbar(sizes, mean_ces, std_ces)
```

```python
!pwd
def graph_hists_overlay(path,path2, col, title):
  fnames=os.listdir(path)
  fnames2 = os.listdir(path2)
  num_hists = len(fnames)
  plt.tight_layout()
  plt.rcParams.update({'font.size': 20})
  fig = plt.figure(figsize=(18,11))

  fnames=sorted(fnames, key=lambda x: int(x.split("_")[1]))
  fnames2=sorted(fnames2, key=lambda x: int(x.split("_")[1]))

  for i, fname in enumerate(fnames):
    fields = fname.split("_")
    size = fields[1]
    data = pd.read_csv(path+fname).dropna()
    data2 = pd.read_csv(path2+[i for i in fnames2 if i.split("_")[1]==size][0])
    #print(data,data2)
    plt.subplot(2,2,i+1)
    plt.hist(data[col], color="r", alpha=0.3, bins = [0.1*i for i in range(15)], label="GPT")
    plt.hist(data2[col], color="b", alpha=0.3, bins = [0.1*i for i in range(15)], label="H3")
    plt.xlabel("Crossentropy")
    plt.ylabel("Frequency")
    plt.title("%s-shot" % size)
    plt.legend()
  
  fig.suptitle(title)
graph_hists_overlay("./","../Noise Results/", " Cross-entropy", "Comparison of Crossentropy Distributions under Noise")
```
